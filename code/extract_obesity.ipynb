```python
import pandas as pd
```

```python
life_df = pd.read_excel('../resources/IHME_county_data_LifeExpectancy_Obesity_PhysicalActivity_NATIONAL.xlsx',sheet_name=0)

life_df
```

```python
life_df = life_df.drop(['Difference in male life expectancy, 1985-2010 (years)','Difference in female life expectancy, 1985-2010 (years)'],axis=1)
life_df
```

```python
life_df = life_df.rename(columns={'Male life expectancy, 1985 (years)':'male_life_expectancy_1985_years',\
                                 'Female life expectancy, 1985 (years)':'female_life_expectancy_1985_years',
                                 'Male life expectancy, 1990 (years)':'male_life_expectancy_1990_years',\
                                 'Female life expectancy, 1990 (years)':'female_life_expectancy_1990_years',\
                                 'Male life expectancy, 1995 (years)':'male_life_expectancy_1995_years',\
                                 'Female life expectancy, 1995 (years)':'female_life_expectancy_1995_years',\
                                 'Male life expectancy, 2000 (years)':'male_life_expectancy_2000_years',\
                                 'Female life expectancy, 2000 (years)': 'female_life_expectancy_2000_years',
                                 'Male life expectancy, 2005 (years)':'male_life_expectancy_2005_years',\
                                 'Female life expectancy, 2005 (years)':'female_life_expectancy_2005_years',\
                                 'Male life expectancy, 2010 (years)':'male_life_expectancy_2010_years',\
                                 'Female life expectancy, 2010 (years)':'female_life_expectancy_2010_years',\
                                 'State':'state','County':'county'})
life_df
```

```python
activity_df = pd.read_excel('../resources/IHME_county_data_LifeExpectancy_Obesity_PhysicalActivity_NATIONAL.xlsx',sheet_name=1)
activity_df
```

```python
activity_df = activity_df.drop(['Difference in male sufficient physical activity prevalence, 2001-2009 (percentage points)',\
                               'Difference in female sufficient physical activity prevalence, 2001-2009 (percentage points)',\
                               'Unnamed: 10',\
                               '*Methodological changes in the 2011 BRFSS may affect comparability between the 2011 estimates and estimates from earlier years'], axis=1)
activity_df
```

```python
activity_df = activity_df.rename(columns={'Male sufficient physical activity prevalence, 2001 (%)':'male_sufficient_physical_activity_prevalence_2001_p',\
                                          'Female sufficient physical activity prevalence, 2001 (%)':'female_sufficient_physical_activity_prevalence_2001_p',\
                                          'Male sufficient physical activity  prevalence, 2009 (%)':"male_sufficient_physical_activity_prevalence_2009_p",\
                                          'Female sufficient physical activity  prevalence, 2009 (%)':'female_sufficient_physical_activity_prevalence_2009_p',\
                                          'Male sufficient physical activity  prevalence, 2011* (%)':'male_sufficient_physical_activity_prevalence_2011_p',\
                                          'Female sufficient physical activity  prevalence, 2011* (%)':'female_sufficient_physical_activity_prevalence_2011_p',\
                                         'State':'state','County':'county'})
activity_df                                          
```

```python
obesity_df = pd.read_excel('../resources/IHME_county_data_LifeExpectancy_Obesity_PhysicalActivity_NATIONAL.xlsx',sheet_name=2)

obesity_df
```

```python
obesity_df = obesity_df.drop(['Difference in male obesity prevalence, 2001-2009 (percentage points)',\
                               'Difference in female obesity prevalence, 2001-2009 (percentage points)',\
                               'Unnamed: 10',\
                               '*Methodological changes in the 2011 BRFSS may affect comparability between the 2011 estimates and estimates from earlier years'], axis=1)
obesity_df
```

```python
obesity_df = obesity_df.rename(columns={'Male obesity prevalence, 2001 (%)':'male_obesity_prevalence_2001_p',\
                                        'Female obesity prevalence, 2001 (%)':'female_obesity_prevalence_2001_p',\
                                        'Male obesity prevalence, 2009 (%)':'male_obesity_prevalence_2009_p',\
                                        'Female obesity  prevalence, 2009 (%)':'female_obesity_prevalence_2009_p',\
                                        'Male obesity  prevalence, 2011* (%)':'male_obesity_prevalence_2011_p',\
                                        'Female obesity prevalence, 2011* (%)':'female_obesity_prevalence_2011_p',\
                                       'State':'state','County':'county'})
obesity_df
```

```python
merged_life_activity = pd.merge(life_df,activity_df, how = 'inner', on=['state','county'])
merged_life_activity
```

```python
merged_all = pd.merge(merged_life_activity,obesity_df,how = 'inner', on=['state','county'])
merged_all
```
